When we checked the last 5 records, we can see all the values in the data set is good except the attribute stalk root. It has a value '?' Which means the missing values. So let work on them now.

then we will encode the value based on how much that value varies the other attribute values.

First will check the count of those missing values. If they are less and doesn't impact much, then we can neglect those values or else, we will know more about that data. 

```
mushrooms['stalk-root'].value_counts()
```

As there are a lot of missing values in the data, we will encode the data. That means we will divide the data into two parts based on the missing values. One is the set of missing values, and the other is the set of values that don't have missing values. Then, we will analyze the data(other attribute values) whether there is a change or they are similar. Let's see whether we will get useful information or not.

```
for attr in mushrooms.columns:
    #Format subplots
    fig, ax = plt.subplots(1,2)
    plt.subplots_adjust(right=2)
    
    #Construct values to count in each column
    a=set(mushrooms[mushrooms['stalk-root']=='?'][attr])
    b=set(mushrooms[mushrooms['stalk-root']!='?'][attr])
    c = a.union(b)
    c = np.sort(np.array(list(c)))
    
    #Build each subplot
    sns.countplot(x=mushrooms[mushrooms['stalk-root']=='?'][attr], order=c, ax=ax[0]).set_title('stalk-root == ?')
    sns.countplot(x=mushrooms[mushrooms['stalk-root']!='?'][attr], order=c, ax=ax[1]).set_title('stalk-root != ?')
    
    #Plot the plots
    fig.show()
```

Since every attribute varied. We can't just remove the missing values. Instead, we have to encode the data.we will do that in assignment4 if it is necessary for our problem statement. As of now, let us explore more into the data.

```
# changing the label for each column instead of a letter to a word, just for the sake of visualization. 
stalk_shape              = dict(e='enlarging',t='tapering')
stalk_root               = {'b':'bulbous','c':'club','u':'cup','e':'equal','z':'rhizomorphs','r':'rooted', '?':'missing'}
stalk_surface_above_ring = dict(f='fibrous',y='scaly',k='silky',s='smooth')
stalk_surface_below_ring = dict(f='fibrous',y='scaly',k='silky',s='smooth')
stalk_color_above_ring   = dict(n='brown', b='buff', c='cinnamon', g='gray', o='orange', p='pink', e='red',
                                w='white', y='yellow')
stalk_color_below_ring   = dict(n='brown', b='buff', c='cinnamon', g='gray', o='orange', p='pink', e='red',
                                w='white', y='yellow')

# Map all the new labels to its corresponding column
mushrooms_visual['stalk-shape']              = mushrooms_visual['stalk-shape'].map(stalk_shape)
mushrooms_visual['stalk-root']               = mushrooms_visual['stalk-root'].map(stalk_root)
mushrooms_visual['stalk-surface-above-ring'] = mushrooms_visual['stalk-surface-above-ring'].map(stalk_surface_above_ring)
mushrooms_visual['stalk-surface-below-ring'] = mushrooms_visual['stalk-surface-below-ring'].map(stalk_surface_below_ring)
mushrooms_visual['stalk-color-above-ring']   = mushrooms_visual['stalk-color-above-ring'].map(stalk_color_above_ring)
mushrooms_visual['stalk-color-below-ring']   = mushrooms_visual['stalk-color-below-ring'].map(stalk_color_below_ring)
```

By the graphs above, let us understand the data.

Stalk Root:

1. If the Stalk root is rooted or equal, then mostly they are edible.

Stalk Surface:

```
# changing the label for each column instead of a letter to a word, just for the sake of visualization.

veil_color  = dict(n='brown', o='orange', w='white', y='yellow')
ring_number = dict(n='none', o='one', t='two')
ring_type   = dict(c='cobwebby', e='evanescent', f='flaring', l='large', n='none', p='pendant',
                   s='sheathing', z='zone')
mushrooms_visual['veil-color']  = mushrooms_visual['veil-color'].map(veil_color)
mushrooms_visual['ring-number'] = mushrooms_visual['ring-number'].map(ring_number)
mushrooms_visual['ring-type']   = mushrooms_visual['ring-type'].map(ring_type)
```

Finally, lets check the other misc features.

```
bruises           = dict(t='true', f='false')
odor              = dict(a='almond', l='anise', c='creosote', y='fishy', f='foul', m='musty', n='none', 
                         p='pungent', s='spicy')
spore_print_color = dict(k='black', n='brown', b='buff', h='chocoloate', r='green', o='Orange', u='Purple',
                        w='White', y='Yellow')
population        = dict(a='Abundant', c='Clustered', n='Numerous', s='Scattered', v='Several', y='Solitary')
habitat           = dict(g='Grasses', l='Leaves', m='Meadows', p='Paths', u='Urban', w='Waste', d='Woods')


mushrooms_visual['bruises']             = mushrooms_visual['bruises'].map(bruises)
mushrooms_visual['odor']                = mushrooms_visual['odor'].map(odor)
mushrooms_visual['spore-print-color']   = mushrooms_visual['spore-print-color'].map(spore_print_color)
mushrooms_visual['population']          = mushrooms_visual['population'].map(population)
mushrooms_visual['habitat']             = mushrooms_visual['habitat'].map(habitat)
```

```
plt.figure(figsize=(18, 10))

labels = ['bruises', 'odor', 'spore-print-color', 'population', 'habitat']

for i, label in enumerate(labels):
    plt.subplot(2, 3, i+1)
    sns.countplot(data=mushrooms_visual, x=label, hue='class')
    plt.title(label.capitalize())
    plt.xticks(rotation=40)

mlt.rc('xtick', labelsize=14)
plt.subplots_adjust(wspace=.3, hspace=.4)
```

Spore Print Color:

1. Black or brown colored are most likely edible.

Population:

1. Most of them are edible apart from several categories.

Habitat:
